```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("Diwali Sales Data.csv", encoding='unicode_escape')
```

```python
df
```

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'],axis=1, inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df["Amount"]=df["Amount"].astype("int")
```

```python
df.rename(columns={"Marital_Status":"Marital_Status 0 is No |1 is Yes"},inplace=True)
```

```python
df.head()
```

```python
df[['Age',"Amount",'Age Group']].describe()
```

```python
df
```

```python
gender_bar=sns.countplot(x="Gender",data=df)
for  bar in gender_bar.containers:
    gender_bar.bar_label(bar)
plt.show()
```

```python
var=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by=["Amount"])
var
```

```python
graph1=sns.barplot(y="Amount",x="Gender",data=var,hue="Gender")

```

```python
var1=df.groupby(["Gender"],as_index=False)["Orders"].sum().sort_values(by=["Orders"])
var1
```

```python
label_order_gender=sns.barplot(x="Gender",y="Orders",data=var1)
for bar in label_order_gender.containers:
    label_order_gender.bar_label(bar)
```

```python
age_groups_graph=sns.countplot(x="Age Group",data =df, hue="Gender")
for bar in age_groups_graph.containers:
    age_groups_graph.bar_label(bar)
plt.show()
```

```python
age_group=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by=["Amount"],ascending=False)
```

```python
age_group
```

```python
age_group_graph=sns.barplot(y="Age Group",x="Amount",data=age_group)
```
